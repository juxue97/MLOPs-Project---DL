```python
import pandas as pd
import os
import shutil


destination_dir = "C:\\Users\\jarvi\\Downloads\\face_emotion_multiclass\\"
os.makedirs(destination_dir,exist_ok=True)

source_folder_path = "C:\\Users\\jarvi\\Downloads\\face emotion.v1i.multiclass\\"
dataset = ["train","test","valid"]
dataset_folder_path = source_folder_path+dataset[0]

dataset_suffix = "\\_classes.csv"


df = pd.read_csv(source_folder_path+dataset[0]+dataset_suffix)
df.head()
```

```python
for _,row in df.iterrows():
    filename = row["filename"]

    # Get class label
    class_label = df.columns[1:][row[1:].values.argmax()]

    # Create subsidiretory 
    class_path = os.path.join(destination_dir,class_label)
    os.makedirs(class_path, exist_ok=True)

    # Define source and destination filepath
    source = os.path.join(dataset_folder_path,filename)
    destination = os.path.join(class_path,filename)

    # Move the file
    if os.path.exists(source):
        shutil.move(source,destination)
        print(f"Moved:{filename} -> {destination}/")
    else:
        print(f"file not found: {source}")
     
print("File move completed")
```

#### Making it into an automate function

```python
destination_dir = "C:\\Users\\jarvi\\Downloads\\face_emotion_multiclass\\"
os.makedirs(destination_dir,exist_ok=True)

source_folder_path = "C:\\Users\\jarvi\\Downloads\\face emotion.v1i.multiclass\\"
dataset = ["train","test","valid"]
dataset_suffix = "\\_classes.csv"

for i in range(len(dataset)):
    dataset_folder_path = source_folder_path+dataset[i]
    df = pd.read_csv(dataset_folder_path+dataset_suffix)
    for _,row in df.iterrows():
        filename = row["filename"]

        # Get class label
        class_label = df.columns[1:][row[1:].values.argmax()]

        # Create subsidiretory 
        class_path = os.path.join(destination_dir,class_label)
        os.makedirs(class_path, exist_ok=True)

        # Define source and destination filepath
        source = os.path.join(dataset_folder_path,filename)
        destination = os.path.join(class_path,filename)

        # Move the file
        if os.path.exists(source):
            shutil.move(source,destination)
            print(f"Moved:{filename} -> {destination}/")
        else:
            print(f"file not found: {source}")
        
    print(f"File move completed for {dataset_folder_path}")
print("Completed")
```
